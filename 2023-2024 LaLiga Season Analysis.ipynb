```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LaLiga = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/Laliga_table_2023_24.csv")
LaLiga[['goals_scored', 'goals_conceded']] = LaLiga['scoresStr'].str.split('-', expand=True).astype(int)

plt.figure(figsize=(10, 6))
sns.barplot(x= LaLiga.name, y= LaLiga.pts , hue=LaLiga.name, palette='viridis')

plt.title('2023-2024 LaLiga Table', fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.xlabel('Team', fontsize=12)
plt.ylabel('Pts', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LaLiga = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/Laliga_table_2023_24.csv")
LaLiga[['goals_scored', 'goals_conceded']] = LaLiga['scoresStr'].str.split('-', expand=True).astype(int)

plt.figure(figsize=(12, 7))
barWidth = 0.20
x = range(len(LaLiga["name"]))

plt.bar([pos - barWidth for pos in x], LaLiga["wins"], width=barWidth, color="green", label="Wins")
plt.bar(x, LaLiga["draws"], width=barWidth, color="blue", label="Draws")
plt.bar([pos + barWidth for pos in x], LaLiga["losses"], width=barWidth, color="red", label="Losses")

plt.xticks(x, LaLiga["name"], rotation=45, ha="right")
plt.xlabel("Team")
plt.ylabel("Number of Games")
plt.title("LaLiga 2023-2024 Season: Wins, Draws, and Losses")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LaLiga_Goals = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/player_top_scorers.csv")
LaLiga_Assists = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/player_top_assists.csv")

GA_LaLiga = pd.merge(LaLiga_Goals,LaLiga_Assists, how="left", on = ["Player", "Matches", "Team", "Country", "Minutes"])
GA_LaLiga.drop('Rank_y', axis=1, inplace=True)
GA_LaLiga[["Goals", "Assists"]] = GA_LaLiga[["Goals", "Assists"]].fillna(0)
GA_LaLiga["Total GA"] = ((GA_LaLiga["Goals"]+GA_LaLiga["Assists"]))

top_10_GA = GA_LaLiga.nlargest(10, "Total GA").sort_values(by='Total GA', ascending=False)
top_10_GA['Player_Team'] = GA_LaLiga['Player'] + " (" + GA_LaLiga['Team'] + ")"

plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_GA.Player_Team,y=top_10_GA["Total GA"],hue=top_10_GA.Player_Team, palette="viridis")

plt.title("Top 10 GA(Goals and Assists) Providers LaLiga 2023-2024", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Players", fontsize=12)
plt.ylabel("Total GA", fontsize=12)


plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

LaLiga_Goals = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/player_top_scorers.csv")
LaLiga_Assists = pd.read_csv("C:/Project DATA/Real Madrid 2023-2024 LaLiga Analysis/player_top_assists.csv")

GA_LaLiga = pd.merge(LaLiga_Goals,LaLiga_Assists, how="left", on = ["Player", "Matches", "Team", "Country", "Minutes"])
GA_LaLiga.drop('Rank_y', axis=1, inplace=True)
GA_LaLiga[["Goals", "Assists"]] = GA_LaLiga[["Goals", "Assists"]].fillna(0)
GA_LaLiga["Total GA"] = ((GA_LaLiga["Goals"]+GA_LaLiga["Assists"]))

Real_Madrid_GoalsAssists = GA_LaLiga[GA_LaLiga["Team"] == "Real Madrid" ].head(5).sort_values(by = "Total GA", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=Real_Madrid_GoalsAssists.Player,y=Real_Madrid_GoalsAssists["Total GA"],hue=Real_Madrid_GoalsAssists.Player, palette="viridis")

plt.title("Top 5 GA(Goals and Assists) Providers Real Madrid LaLiga 2023-2024", fontsize=16)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Players", fontsize=12)
plt.ylabel("Total GA", fontsize=12)


plt.tight_layout()
plt.show()
```
